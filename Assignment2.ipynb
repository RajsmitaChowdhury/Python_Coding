##### Question 1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
def myreduce(func_name,iterable):
    iterab = iter(iterable) #Converting iterable to iterator
    a = next(iterab)   #Taking next iteration
    for i in iterab:
        a = func_name(a,i)
    return a
        
```

```python
def example(s,a):
    return s+a

print(myreduce(example,[2,4,6,8]))
```

##### Question 3 Implement a function longestWord() that takes a list of words and returns the longest one.

```python
def longestWord(wordlist):
    for i in range(1,len(wordlist)):
        if len(wordlist[i]) > len(wordlist[i-1]):  #Check if the specific index elements of the list is greater than the previous one
            longword = wordlist[i]
        else:
            longword = wordlist[i-1]
    return longword
```

```python
longestWord(['appl','bac','aeroplane','efhsakgh','dggd'])
```

##### Question 1.1 Write a Python Program(with class concepts) to find the area of the triangle

```python
"""Python program to find the area of a triangle using Heron's formula"""
class triangle:
    def __init__(self,a,b,c):   #Initialize the side lengths of the triangle
        self.a = a
        self.b = b
        self.c = c
        

class area(triangle):
    def calc_area(self):
        s = (self.a + self.b + self.c)/2  #Calculating s (half of the perimeter)
        ar = (s*(s-self.a)*(s-self.b)*(s-self.c)) ** 0.5  #Calculating area of the triangle
        return ar
```

```python
p = area(5,5,5)
```

```python
p.calc_area()
```

##### Question 1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

```python
"""Filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer 
than n"""
def filter_long_words(wordlist,n):
    l = []
    for i in wordlist:
        if len(i) > n:   #Check if length of word is greater than n
            l.append(i)  #If the above condition is true append in the new list
    return l  
```

```python
filter_long_words(['abc','apple','orange','aeroplane'],5)
```

###### Solution 1

```python
"""Python program to map list of words into its corresponding lengths into a list"""
def wordmap(listword):
    return len(listword)  #Return length of word

l=['abc','de','apple']
print(list(map(wordmap,l)))  #Map function to map the above function to a list
```

###### Solution 2

```python
"""Python program to map list of words into its corresponding lengths into a list"""
def wordmap(listword):
    l = []
    for i in listword:
        l.append(len(i))   #Append the length of words in the new list
    return l
```

##### Question 2.2 Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.

```python
"""Python function which takes 1 character and checks if its a vowel or not"""
def charfunc(char):
    uppercase = char.upper()
    if uppercase in ['A','E','I','O','U']:  #Check if the character is a vowel
        return True
    else:
        return False
```
